# 2. Weekly graph question

Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

Both the histogram and the boxplot summarize the distribution of drug prices, but they emphasize different aspects of the data.

**Histogram – pros & cons:**

A histogram shows the full shape of the distribution, making it easy to see that most drug prices cluster at the lower end while a smaller number of drugs are much more expensive (a right-skewed pattern). This helps the administrator understand how “typical” prices compare to the expensive ones. However, histograms depend on bin choices and don’t show exact summary statistics like quartiles or outliers as clearly.

**Boxplot – pros & cons:**

A boxplot provides a compact summary (median, quartiles, and outliers). It quickly shows how high the upper quartile and maximum prices are relative to the rest, which directly answers how expensive drugs compare to typical ones. The downside is that it hides the detailed shape of the distribution, so you can’t see clustering or gaps.

**Which I’d choose:**

For this question—how expensive drugs compare to the rest—I would choose the boxplot, because it clearly highlights the upper quartile and extreme values in relation to the median. If explaining overall pricing patterns to nontechnical staff, I might show both together.

**Formatting improvements:**

*   Add clear titles and axis labels (e.g., “Distribution of Drug Prices ($)”).
*   Show the 75th-percentile line on the histogram to connect both views.


*   Use consistent scales (same x-axis range).

*   Annotate median and upper-quartile values on the boxplot.

*   Consider a log-scale x-axis (prices are highly skewed), which would spread out lower prices and make differences among expensive drugs clearer.


Overall, combining a labeled boxplot with a log-scaled histogram would make the analysis more visually clear and informative.

```python
# Histograms (Numerical Columns)

df_ecommerce.hist(figsize=(12,8), bins=20)
plt.tight_layout()
plt.show()
```

```python
# KDE Plots

for col in df_ecommerce.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.kdeplot(df_ecommerce[col], fill=True)
    plt.title(f"KDE Plot of {col}")
    plt.show()
```

```python
# Violin PLots

for col in df_ecommerce.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.violinplot(y=df_ecommerce[col])
    plt.title(f"Violin Plot of {col}")
    plt.show()
```

```python
# Grouped Histogram For Sales and Category

sns.histplot(
    data=df_ecommerce,
    x="Sales",
    hue="Category",
    bins=30,
    kde=True
)

plt.title("Sales Distribution by Category")
plt.show()
```

```python
# Grouped Histogram For Quantity and Region

sns.histplot(
    data=df_ecommerce,
    x="Quantity",
    hue="Region",
    bins=30,
    kde=True
)

plt.title("Quantity Distribution by Region")
plt.show()
```

```python
# Grouped Histogram For Profit and Category

sns.histplot(
    data=df_ecommerce,
    x="Profit",
    hue="Category",
    bins=30,
    kde=True
)

plt.title("Profit Distribution by Category")
plt.show()
```

```python
# Grouped Histogram For Profit and Region

sns.histplot(
    data=df_ecommerce,
    x="Profit",
    hue="Region",
    bins=10,
    kde=True
)

plt.title("Profit Distribution by Region")
plt.show()
```

```python


# Survey Data (approximate values to match book example)
categories = ["Bored", "Not great", "OK", "Kind of interested", "Excited"]
before = np.array([11, 5, 40, 25, 19])
after = np.array([12, 6, 14, 30, 38])

# Create stacked horizontal bars
fig, ax = plt.subplots(figsize=(10, 4))

left_before = 0
left_after = 0

for i in range(len(categories)):
    ax.barh("Before", before[i], left=left_before)
    ax.barh("After", after[i], left=left_after)
    left_before += before[i]
    left_after += after[i]

# Formatting
ax.set_xlim(0, 100)
ax.set_xlabel("Percentage of Students")
ax.set_title("Pilot program was a success\nHow do you feel about science?")
ax.set_xticks([])
ax.set_yticks(["Before", "After"])

plt.show()
```
